```python
temp_list = []

for gu in df_3['구'].unique():
    df_4 = df_3[df_3['구'] == gu].copy()
    df_4['평당 임대료']= df_2.loc[gu, '4분기']
    temp_list.append(df_4)
    
df_5 = pd.concat(temp_list)
cols = ['기준_년_코드', '기준_분기_코드', '구', '동', '서비스_업종_코드_명', '분기당_매출_금액', '평당 임대료']
df_5 = df_5[df_5['기준_년_코드'] == 2021]
df_5['매장 월평균 매출'] = df_5['분기당_매출_금액'] / (df_5['점포수'] * 3)
```

```python
df_5.to_csv('seoul_data/sales_summary_data_20214q.csv')
```
